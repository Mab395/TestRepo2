##### Step 1: Install and Import Required Libraries

### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Drop missing values
df_comp = df.dropna(subset=['ConvertedCompYearly'])

# Step 2: Set plot style
sns.set(style='whitegrid')

# Step 3: Plot histogram and KDE (distribution plot)
plt.figure(figsize=(10, 6))
sns.histplot(df_comp['ConvertedCompYearly'], bins=50, kde=True, color='mediumseagreen')
plt.title(" Distribution of Yearly Compensation (ConvertedCompYearly)")
plt.xlabel("Yearly Compensation (USD)")
plt.ylabel("Number of Respondents")
plt.tight_layout()
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here
# Filter data for full-time employees
full_time_df = df[df['Employment'] == 'Employed, full-time']

# Drop missing values in the compensation column
full_time_df = full_time_df[full_time_df['ConvertedCompYearly'].notna()]

# Calculate the median compensation
median_compensation = full_time_df['ConvertedCompYearly'].median()

print("Median Yearly Compensation for Full-Time Employees: ${:,.2f}".format(median_compensation))
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out missing values
filtered_df = df[['Country', 'ConvertedCompYearly']].dropna()

# Optional: Filter countries with enough data (e.g., at least 100 responses)
country_counts = filtered_df['Country'].value_counts()
valid_countries = country_counts[country_counts >= 100].index
filtered_df = filtered_df[filtered_df['Country'].isin(valid_countries)]

# Set up figure size
plt.figure(figsize=(16, 8))

# Create boxplot
sns.boxplot(x='Country', y='ConvertedCompYearly', data=filtered_df)
plt.xticks(rotation=45, ha='right')
plt.title('Yearly Compensation Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Yearly Compensation (USD)')
plt.tight_layout()
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here
# Step 1: Drop NA values in compensation column
comp_df = df[['ConvertedCompYearly']].dropna()

# Step 2: Calculate Q1, Q3, and IQR
Q1 = comp_df['ConvertedCompYearly'].quantile(0.25)
Q3 = comp_df['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1

# Step 3: Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 4: Filter out outliers
df_no_outliers = df[(df['ConvertedCompYearly'] >= lower_bound) & (df['ConvertedCompYearly'] <= upper_bound)]

# Step 5: Print summary
print("Original dataset size:", df.shape)
print("New dataset size after removing outliers:", df_no_outliers.shape)
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Select relevant columns and drop missing values
corr_df = df_no_outliers[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']].copy()
corr_df = corr_df.dropna()

# Step 2: Ensure numeric types
corr_df['WorkExp'] = pd.to_numeric(corr_df['WorkExp'], errors='coerce')
corr_df['JobSatPoints_1'] = pd.to_numeric(corr_df['JobSatPoints_1'], errors='coerce')

# Step 3: Drop rows with any non-numeric issues
corr_df = corr_df.dropna()

# Step 4: Compute correlation matrix
correlation_matrix = corr_df.corr(method='pearson')

# Step 5: Visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f", square=True, linewidths=0.5)
plt.title('Correlation Between Compensation, Experience, and Satisfaction')
plt.tight_layout()
plt.show()
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Filter and clean data
scatter_df = df_no_outliers[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']].copy()
scatter_df = scatter_df.dropna()
scatter_df['WorkExp'] = pd.to_numeric(scatter_df['WorkExp'], errors='coerce')
scatter_df['JobSatPoints_1'] = pd.to_numeric(scatter_df['JobSatPoints_1'], errors='coerce')
scatter_df = scatter_df.dropna()

# Step 2: Create scatter plots
plt.figure(figsize=(14, 6))

# Plot 1: Compensation vs Work Experience
plt.subplot(1, 2, 1)
sns.scatterplot(x='WorkExp', y='ConvertedCompYearly', data=scatter_df, alpha=0.6)
plt.title('Compensation vs Work Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Yearly Compensation (USD)')

# Plot 2: Compensation vs Job Satisfaction
plt.subplot(1, 2, 2)
sns.scatterplot(x='JobSatPoints_1', y='ConvertedCompYearly', data=scatter_df, alpha=0.6)
plt.title('Compensation vs Job Satisfaction')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Yearly Compensation (USD)')

plt.tight_layout()
plt.show()
```
